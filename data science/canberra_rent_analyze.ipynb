```python
def fetch_coordinates(address):
    base_url='https://maps.googleapis.com/maps/api/geocode/json?address='
    url=base_url+address
    response=requests.get(url)
    result_json=response.json()
    try:
        x=result_json['results'][0]['geometry']['location']['lat']
        y=result_json['results'][0]['geometry']['location']['lng']
        #return [x,y]
        print('%f;%f'%(x,y))
        return "%f;%f"%(x,y)
    except:
        print('null')
        return 'null'
```

```python
address=df['address']
address.name='address'
```

```python
def parse_rent(line):
    pattern=re.compile(r'[0-9,]+')
    result=pattern.search(line)
    if result:
        return result.group().replace(',','')
    else:
        return '0'
```

```python
rent=df['rent']
rent=rent.apply(parse_rent)
rent.name='rent'
```

```python
suburb=df['suburb']
suburb.name='suburb'
```

```python
def parse_postcode(line):
    pattern=re.compile(r'\d{4}')
    result=pattern.search(line)
    if result:
        return result.group()
    else:
        return 'n/a'
```

```python
postcode=suburb.apply(parse_postcode)
```

```python
postcode.name='postcode'
```
